```python
import numpy as np
import pandas as pd
import scipy.stats as s
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.impute import SimpleImputer
from Gaussian_Naive_Bayes import GaussianNB
```

```python
train_data=pd.read_csv('scania/aps_failure_training_set.csv')
```

```python
train_data
```

```python
train_data.info
```

```python
train_data.describe(include='all')
```

```python
train_data.replace(to_replace='na',value=np.nan,inplace=True)
```

```python
train_data
```

```python
train_data.dropna(axis=1,inplace=True,thresh=int(0.8*train_data.shape[0]))
```

```python
train_data
```

```python
imputer=SimpleImputer()
```

```python
train_data_array=imputer.fit_transform(X=train_data.iloc[:,1:])
```

```python
data_columns=train_data.columns
train_data_labels=train_data['class']
```

```python
train_data=pd.DataFrame(data=train_data_array,columns=data_columns[1:])
```

```python
train_data
```

```python
train_data['class']=train_data_labels
```

```python
train_data
```

```python
plt.figure(figsize=(20,20))
sns.displot(data=train_data['aa_000'],height=5,aspect=2)
```

```python
plt.hist(train_data['ac_000'],bins=10)
plt.xlabel('ac_000')
plt.ylabel('Frequency')
```

```python
pd.qcut(train_data['aa_000'],q=10,duplicates='drop').cat.codes
```

```python
for column in train_data.columns[0:train_data.shape[1]-1]:
    train_data[column]=pd.qcut(x=train_data[column],q=10,duplicates='drop').cat.codes
```

```python
train_data
```

```python
np_array_list=list()
for column in train_data.columns[0:train_data.shape[1]-1]:
    np_array_list.append(np.eye(10,10)[train_data[column]])
data_array=np.concatenate(np_array_list,axis=1)    
```

```python
data_array.shape
```

```python
train_data=pd.DataFrame(data=data_array)
train_data['class']=train_data_labels
```

```python
train_data
```

```python
obj=GaussianNB(features=train_data.iloc[:,0:train_data.shape[1]-1],labels=train_data['class'],
               data_split_ratio=(0.8,0.2,0.0),apply_pca=True,n_components=100)
```

```python
obj.X_new
```

```python
train_data_new=pd.DataFrame(data=obj.X_new)
train_data_new['class']=train_data_labels
```

```python
train_data_new
```

```python
from imblearn.over_sampling import SMOTE
```

```python
smote_obj=SMOTE(sampling_strategy='minority',k_neighbors=5)
```

```python
x_oversampled,y_oversampled=smote_obj.fit_resample(X=obj.X_new,y=train_data_new['class'])
```

```python
train_data_new=pd.DataFrame(data=x_oversampled)
train_data_new['class']=y_oversampled
```

```python
train_data_new
```

```python
x_train_pos=train_data_new[train_data_new['class']=='pos'].iloc[0:int(0.8*train_data_new.shape[0])//2]
x_train_neg=train_data_new[train_data_new['class']=='neg'].iloc[0:int(0.8*train_data_new.shape[0])//2]
x_cv_pos=train_data_new[train_data_new['class']=='pos'].iloc[int(0.8*train_data_new.shape[0])//2:]
x_cv_neg=train_data_new[train_data_new['class']=='neg'].iloc[int(0.8*train_data_new.shape[0])//2:]
train_data_new=pd.concat([x_train_pos,x_train_neg])
cv_data_new=pd.concat([x_cv_pos,x_cv_neg])
```

```python
train_data_new
```

```python
cv_data_new.shape
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
gaussian_nb=GaussianNB()
```

```python
gaussian_nb.fit(X=np.array(train_data_new.iloc[:,0:train_data_new.shape[1]-1]),y=np.array(train_data_new['class']))
```

```python
predicted_labels=gaussian_nb.predict(X=np.array(cv_data_new.iloc[:,0:cv_data_new.shape[1]-1]))
```

```python
predicted_labels
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_true=np.array(cv_data_new['class']),y_pred=predicted_labels))
```
